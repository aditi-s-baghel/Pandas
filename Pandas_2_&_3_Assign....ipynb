# Q1. Display First Rows

## Problem Statement
Create the following DataFrame:

| Name | Age |
|------|-----|
| Aman | 21 |
| Rahul | 22 |
| Neha | 20 |
| Priya | 23 |
| Ravi | 24 |

## Sample Output
```python

```


# Q8. Handle Missing Values

## Problem Statement
Create a DataFrame:

```python
Name = ["Aman", None, "Neha"]
Marks = [80, 75, None]
```

Fill missing values using `fillna()`.

## Sample Output
```python
Missing values replaced
```


````python
import pandas as pd
data = {
    "Name" : ['Aadi',None,'Neha'],
    "Marks" : [80,75, None]
}
df = pd.DataFrame(data)
df.fillna({
    'Name' : 'Unknown',
    'Marks'  : 0})
````


# Q9. Replace Values

## Problem Statement
Create a DataFrame:

```python
City = ["BPL", "IND", "BPL"]
```

Replace:
- BPL → Bhopal
- IND → Indore

## Sample Output
```python
Bhopal
Indore
Bhopal
```


````python
import pandas as pd
data = {
    'City' : ["BPL","IND","BPL"]
}
df = pd.DataFrame(data)
df.replace({
    "BPL":"Bhopal",
    "IND":"Indore"
},inplace = True)
````
