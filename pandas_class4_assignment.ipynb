# Q1. Select Single Column

## Problem Statement
Create a DataFrame and display only the Name column.

## Sample Input
```python
Name, Age
```

## Sample Output
```python
Only Name column
```


````python
import pandas as pd
data = {
    "Name" : ["Aditi","Aadi","Rishi","Ansh"],
    "Age" : [18,22,24,21]
}
df = pd.DataFrame(data)
print(df['Name'])
````


# Q9. Group By City

## Problem Statement
Find average marks city-wise.

## Sample Input
```python
City and Marks
```

## Sample Output
```python
Grouped result
```


````python
import pandas as pd
data = {
    "City" : ["Bhopal","Delhi","Bhopal","Pune"],
    "Marks" : [90,66,96,72]
}
df = pd.DataFrame(data)

df.groupby('City')['Marks'].mean()
````


# Q10. Multiple Conditions

## Problem Statement
Display students with Marks > 70 and Age > 20.

## Sample Input
```python
Marks and Age
```

## Sample Output
```python
Filtered rows
```


````python
import pandas as pd
data = {
    "Name" : ["Aditi","Aadi","Rishi","Ansh"],
    "Age" : [18,22,19,21],
     "Marks" : [90,66,96,72]   
}
df = pd.DataFrame(data)

df[(df['Marks'] > 70) & (df['Age'] > 20)]
````
